##### recorded sentences between the actors

```python
sentence_recorded_by_both = set(male_actor_clips.original_sentence_id).intersection(set(female_actor_clips.original_sentence_id))
sentence_recorded_only_by_male = set(male_actor_clips.original_sentence_id).difference(set(female_actor_clips.original_sentence_id))
sentence_recorded_only_by_female = set(female_actor_clips.original_sentence_id).difference(set(male_actor_clips.original_sentence_id))
sentence_recorded = set(male_actor_clips.original_sentence_id).union(set(female_actor_clips.original_sentence_id))
sentence_not_recorded = set(sentences.id).difference(set(clips.original_sentence_id))
```

```python
len(sentence_recorded_by_both)
```

```python
data_for_graph = {"sentence_recorded_by_both":len(sentence_recorded_by_both),
                  "sentence_recorded_only_by_male":len(sentence_recorded_only_by_male),
                  "sentence_recorded_only_by_female":len(sentence_recorded_only_by_female),
                  "sentence_recorded":len(sentence_recorded),
                  "sentence_not_recorded":len(sentence_not_recorded),
                  "total sentences":len(sentences),
                 }
data_for_graph = dict(sorted(data_for_graph.items(), key=lambda item: item[1]))
```

```python
clips.loc[:,"caracter_count"] = clips.sentence.str.len() 
male_actor_clips.loc[:,"caracter_count"] = male_actor_clips.sentence.str.len() 
female_actor_clips.loc[:,"caracter_count"] = female_actor_clips.sentence.str.len() 
```

```python
fig,axes = plt.subplots(1,3, figsize=(25,5))
axes[0].bar(clips["caracter_count"].value_counts(ascending=False).index, clips["caracter_count"].value_counts(ascending=False))
axes[0].set_title("All data : caracter_count")
axes[1].bar(male_actor_clips["caracter_count"].value_counts(ascending=False).index, male_actor_clips["caracter_count"].value_counts(ascending=False))
axes[1].set_title("Male Actor : caracter_count")
axes[2].bar(female_actor_clips["caracter_count"].value_counts(ascending=False).index, female_actor_clips["caracter_count"].value_counts(ascending=False))
_=axes[2].set_title("Female Actor : caracter_count")
```

### Create train/vlidation/test set

```python
from sklearn.model_selection import StratifiedKFold
```

```python
skf = StratifiedKFold(n_splits=5, random_state=42, shuffle=True)
for train_index, test_index in skf.split(male_actor_clips, male_actor_clips.caracter_count):
    male_actor_clips_train_with_valid, male_actor_clips_test = male_actor_clips.iloc[train_index], male_actor_clips.iloc[test_index]
    break
skf = StratifiedKFold(n_splits=8, random_state=42, shuffle=True)
for train_index, test_index in skf.split(male_actor_clips_train_with_valid, male_actor_clips_train_with_valid.caracter_count):
    male_actor_clips_train, male_actor_clips_validation = male_actor_clips_train_with_valid.iloc[train_index], male_actor_clips_train_with_valid.iloc[test_index]
    break
male_actor_clips_train.shape, male_actor_clips_test.shape, male_actor_clips_validation.shape
```

```python
male_actor_clips_train.to_csv("male_train.tsv", sep="\t", index=False)
male_actor_clips_test.to_csv("male_test.tsv", sep="\t", index=False)
male_actor_clips_validation.to_csv("male_validation.tsv", sep="\t", index=False)
```

```python
skf = StratifiedKFold(n_splits=5, random_state=42, shuffle=True)
for train_index, test_index in skf.split(female_actor_clips, female_actor_clips.caracter_count):
    female_actor_clips_train_with_valid, female_actor_clips_test = female_actor_clips.iloc[train_index], female_actor_clips.iloc[test_index]
    break
skf = StratifiedKFold(n_splits=8, random_state=42, shuffle=True)
for train_index, test_index in skf.split(female_actor_clips_train_with_valid, female_actor_clips_train_with_valid.caracter_count):
    female_actor_clips_train, female_actor_clips_validation = female_actor_clips_train_with_valid.iloc[train_index], female_actor_clips_train_with_valid.iloc[test_index]
    break
female_actor_clips_train.shape, female_actor_clips_test.shape, female_actor_clips_validation.shape
```

```python
female_actor_clips_train.to_csv("female_train.tsv", sep="\t", index=False)
female_actor_clips_test.to_csv("female_test.tsv", sep="\t", index=False)
female_actor_clips_validation.to_csv("female_valdiation.tsv", sep="\t", index=False)
```
